# IMPUTACIÓN DE DATOS EN CLIENTES.XLSX


```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

```

```python
# Cargamos el datafreame
client = pd.read_excel('clientes.xlsx')
client.head(7)
```

```python
#identificamos los valores nulos
msno.matrix(client)
```

```python
#Cuanttficamos los valores nulos
client.isnull().sum()
```

Se imputarán el nombnre ¨no disponible¨ en los lugares con datos nulos ya que al ser una columna con datos string no es posible, imputar con ayuda de alguna herramienta estadística, sin embargo, para no eliminar las columnas ya que esta informacion es necesaria a final de cuentas


```python
#Imputamos una un string en los valores nulos de la columna 'RFC'
client['RFC'].fillna('No disponible', inplace=True)
```

```python
#Imputamos el mismo string en la columna 'NOMBRE'
client['NOMBRE'].fillna('No disponible', inplace=True)
```

```python
#Verificamos que ya no hay valores nulos
client.isnull().sum()
```

```python
#visualizamos los datos imputados
msno.matrix(client)
```

```python
#Exportamos el dataframe limpio en scv
client.to_csv('clientes_limpio.csv', index=False)
```
