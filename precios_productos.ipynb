# IMPUTACIÓN DE DATOS EN PRECIOS PRODUCTOS.XLSX


```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

```

```python
# Cargamos el datafreame
df = pd.read_excel('precios_productos.xlsx')
df.head(10)
```

```python
df.info()
```

```python
#identificamos los valores nulos
msno.matrix(df)
```

```python
#Cuanttficamos los valores nulos
df.isnull().sum()
```

Se imputa un "no disponible" en la columna NOMBRE_VENDEDOR ya que no es posible acceder a esa información en este momento


```python
#Se imputa 'no disponible' en la columna NOMBRE_VENDEDOR a los valores nulos
df['NOMBRE_VENDEDOR'] = df['NOMBRE_VENDEDOR'].fillna('No disponible')
```

```python
#Verificamos que ya no hay valores nulos
df.isnull().sum()
```

```python
#visualizamos los datos imputados
msno.matrix(df)
```

```python
#Exportamos el dataframe limpio en scv
df.to_csv('precios_productos_limpio.csv', index=False)
```
